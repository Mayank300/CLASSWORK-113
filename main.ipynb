```python
# for bar graph.

import csv
import plotly.graph_objects as go

with open('data.csv', newline='') as f:
  reader = csv.reader(f)
  savings_data = list(reader)

savings_data.pop(0)

total_entries = len(savings_data)
total_people_given_remainder = 0
r = total_people_given_remainder



for data in savings_data:
  if int(data[3]) == 1:
    r+= 1

fig = go.Figure(go.Bar(x=['Remainder is On','Remainder is Off'], y=[r, (total_entries - r)]))
fig.show()

```

```python
# for mean, median, mode & std.

import csv
import plotly.graph_objects as go
import statistics as st

with open('data.csv', newline='') as f:
  reader = csv.reader(f)
  savings_data = list(reader)

savings_data.pop(0)

total_entries = len(savings_data)
total_people_given_remainder = 0
r = total_people_given_remainder

all_savings = []


for data in savings_data:
  all_savings.append(float(data[0]))

  if int(data[3]) == 1:
    r+= 1

print("mean of all_savings -->", st.mean(all_savings))
print("median of all_savings -->", st.median(all_savings))
print("mode of all_savings -->", st.mode(all_savings))
print("stdev of all_savings -->", st.stdev(all_savings))

```
